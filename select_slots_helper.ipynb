```python
import numpy as np
import math
import pandas as pd 
import pickle
import os
from IPython.display import display, Markdown

pd.set_option('display.max_colwidth', None)
```

Set your priorities straight & select a slot to rehearse for the most important rehearsal

```python
priority = ['Choreo', 'Director', 'MD'] # priority you'd like rehearsal times to be assigned (e.g. if Choreo is first entry then that is first rehearsal chosen)
selected_slot = 46 # set number of selected priority one slot (e.g. if choosing School song, then this is {46})
```

Have you run the song_schedule_script.py yet? If not set the below to False and provide the name of the input and output files (don't include the filetype ending for output file please) for that script

```python
run_song_schedule = True
if run_song_schedule:
    # inputs if you've already run the script
    inpickle = 'song_pairings.p' #name of song pairing pickle file (NOT EXCEL SPREADSHEET)
else:
    infile = 'master_timings_list.csv'
    outfile = 'song_pairings1'
    inpickle = f"{outfile}.p"
    os.system(f"python song_schedule_script.py {infile} {outfile}")
```

```python
data = pickle.load(open(inpickle, 'rb'))

pair_mappings = {}
for lead in priority:
    keys = [str(x) for x in data[lead].pair_indices.values]
    pair_mappings[lead] = dict(zip(keys, data[lead].pair_indices.index))


selected_slot_row = data[priority[0]][data[priority[0]].pair_indices == {selected_slot}]
available_pairs = []
for lead in priority[1:]:
    poss_pairs = selected_slot_row.loc[selected_slot_row.index[0], f"{lead}_possible_pairs"]
    poss_pairs_indices = [pair_mappings[lead][str(x)] for x in poss_pairs]
    available_pairs.append(poss_pairs_indices)

display(Markdown(f'## {priority[0]} slot chosen'))
display(selected_slot_row[['pair_indices', 'slot_size', 'cast', 'songs', 'casts', 'slots',
       'cast_sizes', 'cast_overlap']])

i=1
lead = priority[i]
display(Markdown(f"## {lead} slots available"))

lead2_data = data[lead][data[lead].index.isin(available_pairs[i-1])]
display(lead2_data[['pair_indices', 'slot_size', 'cast', 'songs', 'casts', 'slots',
       'cast_sizes', 'cast_overlap']])
```

```python
selected_slot_indices = [7, 10, 11, 12] # choose the slots you'd like to consider for the 2nd priority slot, so the tool can offer 3rd priority slot suggestions
```

```python
i=2
lead=priority[i]
display(Markdown(f"## {lead} slots:"))
lead3_data = data[lead][data[lead].index.isin(available_pairs[i-1])]

lead3_available_pairs = []
for slot_index in selected_slot_indices:
    lead3_available_pairs = [pair_mappings[lead][str(x)] for x in lead2_data.loc[slot_index, f"{lead}_possible_pairs"]]
    display(Markdown(f"#### If slot {slot_index} chosen for {priority[i-1]}, then available MD slots are:"))
    display(lead3_data[lead3_data.index.isin(lead3_available_pairs)][['pair_indices', 'slot_size', 'cast', 'songs', 'casts', 'slots',
       'cast_sizes', 'cast_overlap']])
```
